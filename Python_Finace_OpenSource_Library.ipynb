```python
import yfinance as yf
```

```python
yf.download('AMZN',
            start='2018-01-01',
            end='2018-12-31',
            progress=False,
            auto_adjust=False,
            actions='inline')
```

```python
amzn = yf.Ticker('AMZN')
amzn
```

```python
dir(amzn) # to see object Methods/Attributes
```

```python
amzn.history(period='6mo', interval='5d')
```

```python
amzn.info # info is attribute
```

```python
amzn.get_balancesheet() # 재무재표
```
