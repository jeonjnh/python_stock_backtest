## 1.1 기하평균

```python
data.Change.loc['2021-06-01' : '2021-06-30'].mean() # 산술평균
```

```python
len(data.Change.loc['2021-06-01' : '2021-06-30'])
```

```python
(data.Change.loc['2021-06-01' : '2021-06-30'] + 1).prod() ** (1/len(data.Change.loc['2021-06-01' : '2021-06-30'])) - 1
```

```python
def geo_mean(x):
  return (x + 1).prod() ** (1/len(x)) - 1
```

```python
geo_mean(data.Change)
```

## 4.1 scipy

```python
import scipy as sp
```

```python
data
```

```python
data.reset_index()
```

```python
data = data.reset_index()[['Close']]
data
```

```python
slope, intercept, rvalue, _, _, = sp.stats.linregress(data.index, data.Close)
```

```python
print(slope, intercept)
```

```python
plt.plot(data.index, data.Close)
plt.plot(data.index, data.index*slope + intercept)
plt.show()
```

## scikit-learn

```python
from sklearn.linear_model import LinearRegression
```

```python
data
```

```python
import statsmodels.api as sm
```

```python
index = sm.add_constant(data.index)
index
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(index, data.Close)
```

```python
model.coef_, model.intercept_
```

```python
plt.plot(data.index, data.Close)
plt.plot(index, index*model.coef_ + model.intercept_)
plt.show()
```

## 4.3 statsmodels

```python
index = sm.add_constant(data.index)
model = sm.OLS(data.Close, index)
result = model.fit()
result.summary()
```

```python
result.params
```

```python
result.rsquared
```

```python
plt.plot(data.index, data.Close)
result.fittedvalues.plot()
plt.show()
```
